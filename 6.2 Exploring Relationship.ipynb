```python
# Group the DataFrame by 'zipcode' and find the minimum and maximum prices within each group
result = df.groupby('zipcode')['price'].agg(['min', 'max','mean']).astype(int)
```

```python
result
```

```python
# Calculate the average price based on the number of bedrooms
average_price_by_bedrooms = df.groupby('zipcode')['price'].mean()

print(average_price_by_bedrooms)
```

```python
# Calculate the correlation between 'zipcode' and 'price'
correlation = df['zipcode'].corr(df['price'])
```

```python
correlation
```

```python
#calculate the correlation matrix 
corr_matrix = df.corr()
price_corr = corr_matrix['price'].sort_values(ascending=False)
price_corr
```

```python

# Create a heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')#, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()

# Display the sorted correlations of the 'price' column
print(price_corr)

```

```python

# Select the top variables with the strongest correlations
strongest_corr = price_corr.head().index  # Choose the top 3 correlated variables

# Create scatterplots for the selected variables and identify the variables with the strongest correlations to 'price'.
for var in strongest_corr:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=var, y='price')
    plt.title(f'Scatterplot of variables vs. Price')
    plt.xlabel(var)
    plt.ylabel('Price')
    plt.show()
```

We have scatterplots for the variables with the strongest correlations with 'price'. The variables are related to the 'price' variable. Price are increasing based on the variables sqft_living15,sqft_above,grade,sqft_living.
